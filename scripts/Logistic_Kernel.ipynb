```python
### !!! Your code here !!!!
def sigmoid(x):
    return 1/(1+np.exp(-x))

def klrloss(a,K,Y,lmbda):
    ## !!Your code here!!
    loss=0
    
    for i in range(0,K.shape[0]):
        loss+=-np.log(sigmoid(Y[i]*(K[:,i].T@a))+0.0000000000000001)
    loss+=lmbda*((a.T@K@a)[0][0])/2
    return loss

def klrgrad(a,K,Y,lmbda):
    ## !!Your code here!!
    
    ## compute P=X*w.*Y
    P=sigmoid(np.multiply(np.matmul(K.T,a),Y))
    
    ## grad=-X^T*((1-P).*Y)  (first part)
    grad=-np.matmul(K.T,np.multiply(1-P,Y))+lmbda*(K@a)
    
    ## grad(1:end)+=\lambda*w(1:end)  (second part)
    #grad[1:]+=lmbda*w[1:]
    
    return grad

def debuger(w,eta,newf):
    print(eta,newf)
def trainklr(K,Y,lmbda):
    a0 = np.random.randn(K.shape[1],1)
    return graddesc(a0,
                   lambda a:klrloss(a,K,Y,lmbda),
                   lambda a:klrgrad(a,K,Y,lmbda),
                   ittfn=None)
# provided code for grad desc. eta is starting value. ittfn is optional function to call on each iteration
def graddesc(w,fn,gradfn,eta = 0.3, ittfn=None):
    oldf = fn(w)
    df = 1
    while(df>1e-6):
        g = gradfn(w)
        eta=0.3
        while eta>1e-10:
            neww = w - eta*g
            newf = fn(neww)
            if oldf>newf*1.001:
                break
            eta *= 0.5
        if ittfn is not None:
            ittfn(w,eta,newf)
#         if eta<=1e-10:
#             break
        df=newf-oldf
        oldf = newf
        
        w = neww
    return w
# def graddesc(w,fn,gradfn,eta=0.1,ittfn=None):
#     oldf = fn(w)
#     df = 1
#     while(df>1e-6):
#         w = w - eta*gradfn(w)
#         newf = fn(w)
#         df = oldf-newf # hope to be positive, or we've over-shot and will be done
#         if ittfn is not None:
#             ittfn(w,eta,newf)
#         oldf = newf
#     return w

def up(x):
    if len(x.shape)==1:
        return x[np.newaxis,:]
    else:
        return x

def rbfkernel(x1,x2,sigma):
    x1 = up(x1)
    x2 = up(x2)
#     print(x1.shape,x2.shape)
#     print((x1[:,np.newaxis,:]- x2[np.newaxis,:,:]).shape)
    d = x1[:,np.newaxis,:] - x2[np.newaxis,:,:]
    
    return np.exp(-np.sum(d*d,2)/(2*sigma*sigma))

def klrerrorrate(K,Y,a):
    print(K.shape,a.shape,Y.shape)
    return np.sum(Y*(K.T@a)<=0)/Y.shape[0]
```
